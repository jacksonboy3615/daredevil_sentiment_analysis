```python
import pandas as pd
import numpy as np
```

```python
daredevil_netflix_df = pd.read_csv('daredevil_original_data.csv')

daredevil_born_again_df = pd.read_csv('daredevil_born_again_data.csv')

daredevil_series_full_df = pd.concat([daredevil_netflix_df, daredevil_born_again_df])

daredevil_series_final_df = daredevil_series_full_df.drop('Unnamed: 0', axis=1)
```

```python
daredevil_series_final_df.to_csv('daredevil_drama_full_data.csv')
```

```python
# Sentiment Data

daredevil_original_sentiment_df = pd.read_csv('daredevil_netflix_sentiment.csv')

daredevil_born_again_sentiment_df = pd.read_csv('daredevil_born_again_sentiment.csv')

daredevil_full_sentiment_df = pd.concat([daredevil_original_sentiment_df, daredevil_born_again_sentiment_df])
```

```python
daredevil_full_sentiment_final = daredevil_full_sentiment_df.drop('Unnamed: 0', axis=1)

daredevil_full_sentiment_final
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sentiment_vader = SentimentIntensityAnalyzer()
daredevil_full_sentiment_final['Title_Sentiment_Score'] = daredevil_full_sentiment_final['Titles'].apply(lambda text: sentiment_vader.polarity_scores(text)['compound'])
daredevil_full_sentiment_final['Review_Sentiment_Score'] = daredevil_full_sentiment_final['Reviews'].apply(lambda text: sentiment_vader.polarity_scores(text)['compound'])

def sentiment_category_vader(sentiment):
        if sentiment >= 0.05:
            return "Positive"
        elif sentiment <= -0.05:
            return "Negative"
        else:
            return "Neutral"
        
daredevil_full_sentiment_final['Title_sentiment_Vader'] = daredevil_full_sentiment_final['Title_Sentiment_Score'].apply(sentiment_category_vader)
daredevil_full_sentiment_final['Review_Sentiment_Vader'] = daredevil_full_sentiment_final['Review_Sentiment_Score'].apply(sentiment_category_vader)

daredevil_full_sentiment_final
```

```python
daredevil_full_sentiment_final.to_csv('daredevil_full_sentiment.csv')
```
