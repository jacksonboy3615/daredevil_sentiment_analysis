```python
import pandas as pd
import numpy as np
import requests
# from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
user_rating = []
review_date = []
review_title = []
review = []
likes = []
dislikes = []

url = 'https://www.imdb.com/title/tt18923754/reviews/?ref_=tt_ov_ql_2'
# option = webdriver.ChromeOptions()
# option.add_argument("headless")
driver = webdriver.Chrome()
driver.get(url)

see_all = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section/div/section/div/div[1]/section[1]/div[3]/div/span[2]/button')

while True:
    try:
        see_all.send_keys(Keys.ENTER)
        # time.sleep(120)
    except Exception as e:
        break

for i in range(1,450):
    try:
        rating =  driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section/div/section/div/div[1]/section[1]/article[{}]/div[1]/div[1]/div[1]/span/span[1]'.format(1*i)).text
        date = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section/div/section/div/div[1]/section[1]/article[{}]/div[2]/ul/li[2]'.format(1*i)).text
        title = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section/div/section/div/div[1]/section[1]/article[{}]/div[1]/div[1]/div[2]/div/a/h3'.format(1*i)).text
        review_content = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section/div/section/div/div[1]/section[1]/article[{}]/div[1]/div[1]/div[3]/div/div/div'.format(1*i)).text
        like = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section/div/section/div/div[1]/section[1]/article[{}]/div[1]/div[2]/div[1]/div/span[1]/span[3]'.format(1*i)).text
        dislike = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section/div/section/div/div[1]/section[1]/article[{}]/div[1]/div[2]/div[1]/div/span[2]/span'.format(1*i)).text

        user_rating.append(rating)
        review_date.append(date)
        review_title.append(title)
        review.append(review_content)
        likes.append(like)
        dislikes.append(dislike)

    except:
        continue
```

```python
print(len(user_rating), len(review_date), len(review_title),  len(review), len(likes), len(dislikes))
```

```python
daredevil_born_sentiment_raw = {'User_Ratings': user_rating, 
                           'Review_Dates': review_date,
                           'Titles': review_title,
                           'Reviews': review,
                           'Likes': likes,
                           'Dislikes': dislikes}

daredevil_born_sentiment_df = pd.DataFrame(daredevil_born_sentiment_raw)
```
